```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('C:\\Users\\vrush\\OneDrive\\Desktop\\tennisball.png')
plt.imshow(img,cmap = 'gray')
plt.axis('off')
plt.title('Original Image')
```

```python
img1 = cv2.Canny(img,50,150)
```

```python
 mask = np.zeros_like(img1)
  center = (img1.shape[1] // 2, img1.shape[0] // 2)
 radius = img1.shape[0] // 5  # Adjust radius based on expected ball size
 cv2.circle(mask, center, radius, (255, 255, 255), -1)
 masked_canny = cv2.bitwise_and(img1, mask)
```

```python
 circles = cv2.HoughCircles(masked_canny, cv2.HOUGH_GRADIENT, 1, 20,
                             param1=50, param2=30, minRadius=radius-5, maxRadius=radius+5)
```

```python
plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray'),plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2),plt.imshow(img1,cmap = "gray"),plt.title("Canny")
plt.axis("off")
```
